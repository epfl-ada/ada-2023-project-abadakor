```python
import pandas as pd
```

```python
movies_raw = pd.read_table('movie.metadata.tsv', names=['wikipedia_ID', 'freebase_ID', 'name', 'release_date', 'box_office', 'runtime', 'languages', 'countries', 'genres'])
movies_raw.head()
```

```python
movies_raw.shape
```

```python
ids = pd.read_table('wikidata_freebase_imdb.tsv', names=['wikidata_ID', 'freebase_ID', 'IMDB_ID'], header=0)
ids.head()
```

```python
ids.wikidata_ID = ids.wikidata_ID.str[31:].copy()
```

```python
# Quelles valeurs dupliquées garde-ton ?
ids[ids.duplicated('freebase_ID', keep=False)].sort_values('freebase_ID').iloc[60:120]
```

```python
# Pour l'instant, aucune valeur dupliquée
movies = movies_raw.merge(ids.drop_duplicates(['freebase_ID']), on='freebase_ID', how='left', validate='1:1')
movies.head()
```

```python
print(f'{movies.IMDB_ID.notna().mean()*100:.2f}% of movies in this database have corresponding Wikidata and IMDB pages.')
```
