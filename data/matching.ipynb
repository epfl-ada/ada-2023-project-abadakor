```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import the ratings data
ratings = pd.read_csv('raw_data/imdb_rating/imdb_rating.tsv', sep='\t', header=0)
```

```python
display(ratings)
```

```python
# Plot the distribution of ratings
plt.hist(ratings['numVotes'], bins=20)
plt.xlabel('Rating')
plt.ylabel('Number of movies')
plt.title('Distribution of ratings')

#set y axis to log scale
plt.yscale('log')
plt.show()
```

```python
# Plot the distribution of ratings
plt.hist(ratings['averageRating'], bins=20)
plt.xlabel('Rating')
plt.ylabel('Number of movies')
plt.title('Distribution of ratings')

#set y axis to log scale
# plt.yscale('log')
plt.show()
```

```python
# dentisy plot of ratings vs number of votes
ratings.plot.scatter(x='averageRating', y='numVotes', alpha=0.005)
plt.xlabel('Rating')
plt.ylabel('Number of votes')
plt.title('Number of votes vs rating')
plt.yscale('log')


plt.axhline(y=ratings['numVotes'].mean(), color='red') #mean
plt.axhline(y=ratings['numVotes'].median(), color='black') #median

plt.show()
```
